```python
from DataGeneration import *
from StockExchange import *
```

```python
# Enter values for stock names
stocks = ["TBC ATTO ATH ATKR BEST BGSF BHP MEN MUC MUH MHD CAT CATO CBZ CBL CBO DHX DHT DEO EVT ETO ETG ETB GRUB PAC ASR AVAL SUPV TV GSX"]
stocks = stocks[0].split()

# create a list of 50 random stocks (with replacement) for buy and sell
buy_stocks = random.choices(stocks, k=50)
sell_stocks = random.choices(stocks, k=50)

# create some random buy orders, sell orders and fee ladders
buy_orders = create_orders(buy_stocks, 1001)
sell_orders = create_orders(sell_stocks, 5001)
buy_orders2 = create_orders(buy_stocks, 2001)
sell_orders2 = create_orders(sell_stocks, 6001)
fee_ladder = ((20, 0.1), (40, 0.075), (100, 0.05))
fee_ladder2 = ((30, 0.07), (60, 0.05), (100, 0.02))

# create OrderBook class objects
buy_order_book = OrderBook(buy_orders)
sell_order_book = OrderBook(sell_orders)
buy_order_book2 = OrderBook(buy_orders2)
sell_order_book2 = OrderBook(sell_orders2)

# create Exchange class objects
NYSE = Exchange(buy_order_book, sell_order_book, fee_ladder)
LSE = Exchange(buy_order_book2, sell_order_book2, fee_ladder2)
exchanges = [NYSE, LSE]

# create SORT class object
sort = SORT(exchanges, stocks)

# instantiate UserMenu object
user_menu = UserMenu(NYSE, sort)

# call the main_menu method to open the user menu
user_menu.main_menu()
```

```python
# write the order books to csv files
FileManager.write_to_csv(sort.exchanges[0].buy_order_book.orders, "buy orders.csv")
FileManager.write_to_csv(sort.exchanges[0].sell_order_book.orders, "sell orders.csv")
```
